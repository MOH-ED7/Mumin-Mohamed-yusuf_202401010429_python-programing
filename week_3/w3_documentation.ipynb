## Activity 1: Identify the Components

*   **Inputs:** 
    *   Quiz 1 mark (`quiz_1`)
    *   Quiz 2 mark (`quiz_2`)
    *   Quiz 3 mark (`quiz_3`)
    *   User's decision to continue (`choice`)
*   **Process:**
    *   Prompt and read the three quiz marks.
    *   Calculate the average using the formula: $\text{average} = \frac{\text{quiz\_1} + \text{quiz\_2} + \text{quiz\_3}}{3}$
    *   Evaluate if the average is greater than or equal to 50 to determine a Pass or Fail status.
    *   Loop the entire process as long as the user wants to continue (`choice == 'y'`).
*   **Output:**
    *   The calculated average mark.
    *   The pass/fail status ("Status: Pass" or "Status: Fail").
    *   A "Program Ended" termination message.
    

## Activity 2: Design the Algorithm

### Pseudocode
1. Set choice to "y"
2. While choice is equal to "y":
    * a. Input quiz_1, quiz_2, quiz_3
    * b. Calculate average = (quiz_1 + quiz_2 + quiz_3) / 3
    * c. Display average
    * d. If average >= 50 Then
        * Display "Status: Pass"
    * Else
        * Display "Status: Fail"
    * e. Input choice (Ask to continue)
3. Display "Program Ended"

### Algorithm Structure Identification
*   **Sequence:** Line-by-line execution inside the loop, such as capturing the 3 inputs one after another and computing the `average` calculation.
*   **Selection:** The `if/else` block used to check if `average >= 50` and decide whether to print "Pass" or "Fail".
*   **Iteration:** The `while choice == "y"` loop that repeats the quiz input and calculation process until the user decides to stop.
